### 1장 연습문제 ###


1. 머신러닝을 어떻게 정의할 수 있나요?
명시적인 프로그래밍 없이 컴퓨터가 학습하는 능력을 갖추게 하는 연구 분야다.

2. 머신러닝이 도움을 줄 수 있는 문제 유형 네 가지를 말해보세요.
 기존 솔루션으로는 많은 수동 조정과 규칙이 필요한 문제, 전통적인 방식으로는 해결 방법이 없는 복잡한 문제, 유동적인 환경, 복잡한 문제와 대량의 데이터에서 통찰 얻기


3. 레이블된 훈련 세트란 무엇인가요?
    답이 있는 훈련 세트

4. 가장 널리 사용되는 지도 학습(레이블이 있는 훈련 세트) 작업 두 가지는 무엇인가요?
    분류와 회귀

5. 보편적인 비지도 학습(레이블이 없는 훈련 세트) 작업 네 가지는 무엇인가요?
 계층 군집, 시각화, 차원 축소, 특성 추출, 이상치 탐지, 특이치 탐지, 연관 규칙 학습

6. 사전 정보가 없는 여러 지형에서 로봇을 걸어가게 하려면 어떤 종류의 머신러닝 알고리즘을 사용할 수 있나요?
    강화 학습

7. 고객을 여러 그룹으로 분할하려면 어떤 알고리즘을 사용해야 하나요?
    계층 군집, 분류

8. 스팸 감지의 문제는 지도 학습과 비지도 학습 중 어떤 문제로 볼 수 있나요?
    지도 학습

9. 온라인 학습 시스템이 무엇인가요?
     연속적인 데이터, 빠른 변화에 스스로 적응


10. 외부 메모리 학습이 무엇인가요?
    점진적인 학습 => 컴퓨터가 감당하기 힘든 대용량의 데이터세트를 이용하여 학습함

11. 예측을 하기 위해 유사도 측정에 의존하는 학습 알고리즘은 무엇인가요?
    사례 기반 학습

12. 모델 파라미터와 학습 알고리즘의 하이퍼파라미터 사이에는 어떤 차이가 있나요?
    훈련 중 모델 파라미터는 조정이 가능하고 하이퍼파라미터는 조정이 불가하다.
    알고리즘 애의 파리미터는 모델 파라미터 알고리즘 자체의 파라미터는 하이퍼 파라미터이다.

13. 모델 기반 알고리즘이 찾는 것은 무엇인가요? 성공을 위해 이 알고리즘이 사용하는 가장 일반적인 전략은 무엇인가요? 예측은 어떻게 만드나요?
    새로운 데이터에서 좋은 예측, 미래를 예측한다. 비용 함수를 최소화하는 모델 파라미터를 찾는다.

14. 머신러닝의 주요 도전 과제는 무엇인가요?
    나쁜 알고리즘과 나쁜 데이터를 구분하고 좋은 학습 알고리즘을 선택해서 어떤 좋은 데이터에 훈련시키는 것

15. 모델이 훈련 데이터에서의 성능은 좋지만 새로운 샘플에서의 일반화 성능이 나쁘다면 어떤 문제가 있는 건가요? 가능한 해결책 세 가지는 무엇인가요?
    과대적합, 파라미터 수가 적은 모델을 선택하거나 훈련데이터에 있는 특성을 줄이거나, 모델에 제약을 가하여 단순화 시키기. 훈련 데이터를 더 많이 모으기. 훈련 데이터의 잡음을 줄이기(오류 데이터 수정이나 이상치 제거등)

16. 테스트 세트가 무엇이고 왜 사용해야 하나요?
    새로운 샘플에 모델이 얼마나 잘 작동하는지 확인하기 위해


17. 검증 세트의 목적은 무엇인가요?
    홀드아웃 검증을 사용하여 여러 후보 모델을 평가하고 가장 좋은 하나를 선택하기 위해서

18. 훈련-개발 세트가 무엇인가요? 언제 필요하고 어떻게 사용해야 하나요?
    훈련 데이터가 실제 사용될 데이터를 완벽하게 대표하지 못할 경우에 다른 곳에서 가져온 데이터를 이용하여 또 다른 세트를 만들어 훈련세트에서 훈련시킨 모델을 훈련-개발 세트에서 평가하여 모델이 훈련세트에 과적합 되었는지 확인한다.

19. 테스트 세트를 사용해 하이퍼 파라미터를 튜닝하면 어떤 문제가 생기나요?
    실제 데이터 세트에서 좋은 성능을 기대하기 힘들다.
